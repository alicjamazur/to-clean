```python
# This server accepts POST method and returns a created message introduced to "message" field

from http.server import HTTPServer, BaseHTTPRequestHandler
from urllib.parse import parse_qs

class MessageHandler(BaseHTTPRequestHandler):
    def do_POST(self):
        
        length = int(self.headers.get("Content-Length", 0))    # Indicate the length of the message
        data = self.rfile.read(length).decode()                # Read the correct amount of data from the request.
        message = parse_qs(data)["message"][0]                 # Extract the "message" field from the request data.
        
        self.send_response(200)                                # Send the "message" field back as the response.
        self.send_header('Content-type', 'text/plain; charset=utf-8')
        self.end_headers()
        self.wfile.write(message.encode())

if __name__ == '__main__':
    server_address = ('', 8000)
    httpd = HTTPServer(server_address, MessageHandler)
    httpd.serve_forever()
```
