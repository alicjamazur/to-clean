```python
import pandas as pd 
filename = 'nyc_subway_weather.csv'
subway_df = pd.read_csv(filename)
```

```python
df = subway_df.groupby(['latitude', 'longitude'], as_index=False).mean()
df.head()
```

```python
from bokeh.io import output_notebook, show
from bokeh.models import ColumnDataSource, GMapOptions, ColorBar, BasicTicker
from bokeh.plotting import gmap
from bokeh.palettes import RdYlGn11
from bokeh.models.mappers import LinearColorMapper

map_options = GMapOptions(lat=df['latitude'].mean(), lng=df['longitude'].mean(), map_type="roadmap", zoom=11)
p = gmap("AIzaSyBgXSDXAwnntfyaR6WihXiq-1NB9uozTtI", map_options, title="Subway ridership in NY")

source = ColumnDataSource(data=dict(
                                    latitude=df['latitude'],
                                    longitude=df['longitude'],
                                    size=3*df['ENTRIESn_hourly']/1000,
                                    color=df['ENTRIESn_hourly']))

mapper = LinearColorMapper(palette=RdYlGn11)

color_bar = ColorBar(color_mapper=mapper, ticker=BasicTicker(),
                     label_standoff=12, border_line_color=None, location=(0,0))

p.circle(x="longitude", y="latitude", size="size", fill_color={'field': 'color', 'transform': mapper}, 
         fill_alpha=0.7, line_color=None, source=source)


p.add_layout(color_bar, 'right')
output_notebook()
show(p)
```
